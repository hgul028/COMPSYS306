**Step 6:** Normalization and Standardization of Data:

Often data in it's actual form does not follow any pattern, upper or lower bounds. Therefore, performing operations on such data is difficult. Designing models to work on such data would be even more difficult and inefficient. As such, scaling of data/scaling of features(in the case of images) is important. This is a 'preprocessing step' and you are going to hear this word 'preprocessing' a lot in your Machine Learning journey.  


There are two **different** concepts/techniques to scale data and they are namely, **Normalization and Standardization**. Often they is confused with each other. This should not be done as they are very different.

**Normalization:** Often also called as Min-Max scaling, it is used to transform features to be on a similar scale with uniform lower and upper bounds. The scale often ranges from [0,1] or [-1,1].  Mathematically it is given by,

\begin{align}
X(new) = \frac{X - X(min) \, }{X(max) - X_ (min) \,}
\end{align}

**Standardization:** Often called as Z-score, it is the transformation of the data or features in the data by subtracting mean and dividing by standard deviation. Mathematically it is given by, 

\begin{align}
X(new) = \frac{X - mean () \, }{Standard Deviation () \,}
\end{align}

when we standardize the data the values are centered around the mean with a unit standard deviation.



We will use scikit-learn to normalize and standardize the data in  Iris Flower Database and observe what difference did it make after Normalization and Standardization. 

```python
#printing the first 10 rows of the dataset
import pandas as pd

irisdataset = '/content/gdrive/MyDrive/CS306/database/numerical/IRIS.csv'
df = pd.read_csv(irisdataset) 
#print(df.shape)

first =df.head(10)
print(first.shape)
print(first)
```

We will use a library called seaborn to see two of the columns in the database on a 2D space. This will help understand what Normalization and Standardization does. 

More information about vizualization can be found on Kaggle webpage using the link below. 

https://www.kaggle.com/code/zachgold/python-iris-data-visualizations/notebook 

```python
#graphical representation of the above data  (note: we are only plotting 2 columns )
#we are using a library called seaborn to plot two of the columns sepal_length and sepal_width on a 2-D plane
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x='sepal_length', y='sepal_width', data=df, s=100, color='red')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# labels and title
plt.xlabel('sepal_length', fontsize=14)
plt.ylabel('sepal_width', fontsize=14)
plt.title('sepal_length vs sepal_width ', fontsize=14)
```

It can be seen that the data ( in this case sepal_width and sepal_length) ranges from 2.0 to 4.5 and 4.5 to 8.0 respectively. 

Now we use scikit-learn to normalize the data to bring it within the range [0-1]

```python
#Normalizing the data
#We're using MinMaxScaler() from the sklearn library to normalize the data by scaling it in the range (0,1)
from sklearn.preprocessing import MinMaxScaler
df_n = df.copy()
min_max_scaler = MinMaxScaler()
df_n.iloc[:, [0, 1, 2, 3]] = min_max_scaler.fit_transform(df_n.iloc[:, [0, 1, 2, 3]])
df_n.head(10)
```

```python
#visualizing part of the normalized data using seaborn library 
import seaborn as sns
sns.scatterplot(x='sepal_length', y='sepal_width', data=df, s=100, color='red')
sns.scatterplot(x='sepal_length', y='sepal_width', data=df_n, s=100, color='green')
plt.title('sepal_length vs sepal_width: Comaprison between normal (red) and normalized (green) ', fontsize=14)
```

Now we will perform Standardization on the same data to visualize the normal data vs standardized data for sepal_width and sepal_length.

```python
#Standardization of data 
# apply StandardScaler for iris data set, this is z-score normalization
from sklearn.preprocessing import StandardScaler
df_s = df.copy()
std_scaler = StandardScaler()
df_s.iloc[:, [0, 1, 2, 3]] = std_scaler.fit_transform(df_s.iloc[:, [0, 1, 2, 3]])
df_s.head(10)
```

```python
#Visualization of the standardized data
#normal vs standardized 
sns.scatterplot(x='sepal_length', y='sepal_width', data=df, s=100, color='red')
sns.scatterplot(x='sepal_length', y='sepal_width', data=df_s, s=100, color='blue')
plt.title('sepal_length vs sepal_width: Comaprison between normal (red) and standardized (blue) ', fontsize=14)
```

Now we saw what happens in normalization and standardization of database.
The effect of normalization and standardization can only be seen and best appreciated when a model is designed. We will be applying this concept in future labs.
